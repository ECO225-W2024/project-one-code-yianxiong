## Introduction
**Background Information of the Research**  
Venture capital (VC) is a form of private equity investment where firms or funds provide capital to startups or small businesses with strong growth potential in exchange for equity, or partial ownership. This relationship is significant for startups since they often lacking access to capital markets, because VCs not only inject necessary funds but also offers business expertise, mentorship, and access to a wider network. VC investments are typically made in early-stage companies, aiming to cash out with a significant return through an event like an initial public offering (IPO) or sale (Zider, 1998). The dynamic between VCs and startups is fundamental, with VC acting as a crucial bridge for startups to scale operations, accelerate growth, and realize their market potential. This dynamic not only accelerates technological innovation and economic development but also contributes to job creation and competitive market dynamics. 
  
**Purpose of the Research**  
The investigation into "Exploring Venture Capital Dynamics: The Impact of Pre-Money Valuation and Other Factors on Startup Fundraising Success?" serves a various purpose, which rooted in both the operational strategies of startups and the broader economic implications. This research is aiming for presenting how do factors such as pre-money valuation influences the trajectory of startup fundraising development, and by exploring this relationship, stakeholders across the startup ecosystem can gain valuable insights into the mechanisms of startup growth, investment strategies, and the broader implications for innovation and economic development. From an economic perspective, factors like pre-money valuation and geographic location significantly impact startup fundraising success. This research explores how these variables shape capital flow within the startup ecosystem, influencing growth and expandability. By delving into these dynamics, we aim to provide insights that can guide stakeholders in developing targeted support and strategic interventions to bolster economic development and innovation.  

In the second part of the research, we explore the impact of regional disparities on early-stage funding, drawing on insights from Jun Chen and Michael Ewens (2021). They highlight that areas outside major capital clusters often face financing constraints, leading to missed investment opportunities for high-growth startups. This suggests a potential link between a startup's location and its funding success. Given these regional differences, it's crucial to investigate how geography influences funding dynamics to uncover patterns that could inform investment strategies and regional economic policies.

## 1.1 Variable Explanation
**Outcome(Y):**  
**'raised_amount_usd'** from *"funding_rounds.csv"* is the outcome variable and this numerical variable represents the total amount of money raised by a startup in a specific funding round, denominated in USD. It serves as the primary outcome of interest, quantifying the financial success of fundraising efforts. This variable is central to the research question as it directly quantifies the financial achievements of a startup's fundraising initiatives.

**Independent Variables (X):**   
- **'pre_money_valuation'** from *"funding_rounds.csv"* is a numerical variable representing the valuation of a startup before receiving a new round of investment. It is a critical metric used to determine the equity stake that investors will receive in exchange for their capital. A higher pre-money valuation indicates a higher perceived value of the startup by investors. This variable is essential for understanding the dynamics of venture capital negotiations and the impact of valuation on fundraising success. It reflects investor confidence and the startup's growth prospects, serving as a benchmark for the financial health and potential of the startup company.
  
- **'participants'** from *"funding_rounds.csv"* is a numerical variable indicates the number of investors participating in a given funding round. A higher number of participants can signal strong investor interest and competition, potentially increasing the total raised amount. It reflects the breadth of venture capital support and can be a representation for the size of VC networks and resources available to the startup.
  
- **'fund_raised_amount'** from *"funds.csv"* is a numerical variable measures the amount of capital raised by venture capital funds themselves. Originally called **'raised_amount'**, the name of the varibale will be modified in the 'Data Cleaning' section due to the exsistence of the identical variable name in *"funding_rounds.csv"*. Although this variable focusing on the funds rather than the startups, it can indirectly reflect the investment capacity and aggressiveness of VC firms in the market. Larger fund sizes might indicate a greater willingness or ability to make substantial investments in startups, influencing the outcome of the total amount of money raised by a startup company in a specific funding round.
  
- **'funding_round_code'** from *"funding_rounds.csv"* is a categorical variable categorizes the funding round such as Seed, Series A, Series B, etc. Different funding rounds attract different levels and types of VC interest, the stage of funding can influence both the likelihood and the amount of VC participation. Such that later rounds typically involving larger amounts and more established venture capital firms.  

## 1.2 Data Cleaning

In previous version of the project, the primary focus is on the table of *"funds.csv"* and *"funding_rounds.csv"* and the objective is to combine these two tables into a singular dataset tailored to our analytical needs. 

To prepare for this integration, first eliminate variables that do not contribute to our analysis, such as **'created_at'** and **'updated_at'**. These variables contributes solely to the dataset's management history, and have no relevance to the research objectives. 

```python
import pandas as pd

# Read the file of funds.csv
df = pd.read_csv('funds.csv')
# Drop the specified columns in funds.csv
df = df.drop(['id', 'created_at', 'updated_at'], axis=1)
# Write the modified DataFrame back to funds.csv
df.to_csv('funds.csv', index=False)

# Read the file of funding_rounds.csv
df = pd.read_csv('funding_rounds.csv')
# Drop the specified columns in funding_rounds.csv
df = df.drop(['id', 'created_by', 'created_at', 'updated_at'], axis=1)
# Write the modified DataFrame back to funding_rounds.csv
df.to_csv('funding_rounds.csv', index=False)
```

Additionally, to prevent ambiguity during the analysis, we will rename the variables in *"funds.csv"* which has identical variable names in *"funding_rounds"*. For example, the variable **'funded_at'** in *"funds.csv"* will be renamed to **'fund_funded'**. This is to distinguish different variables from the identical variable names in *"funds.csv"* and *"funding_rounds.csv"*.

```python
# Read the funds.csv file
df = pd.read_csv('funds.csv')

# Rename the column
df = df.rename(columns={'funded_at': 'fund_funded'})
df = df.rename(columns={'object_id': 'fund_object_id'})
df = df.rename(columns={'raised_amount': 'fund_raised_amount'})
df = df.rename(columns={'raised_currency_code': 'fund_raised_currency_code'})
df = df.rename(columns={'source_url': 'fund_source_url'})
df = df.rename(columns={'source_description': 'fund_source_description'})
# Write the modified DataFrame back to the CSV
df.to_csv('funds.csv', index=False)
```

With the above adjustments, we will then merge the modified datasets to form a data foundation for the first phase of the research project.

```python
# Load the datasets
df1 = pd.read_csv('funds.csv')
df2 = pd.read_csv('funding_rounds.csv')

# Combine the DataFrames horizontally
combined_df = pd.concat([df1, df2], axis=1)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('dataset_project_1.csv', index=False)

# Adjust data type for each column according to the  data
column_types = {
    'fund_object_id': object,
    'name': object,
    'fund_funded': object,
    'fund_raised_currency_code': object,
    'fund_source_url': object,
    'fund_source_description': object,
}
df = pd.read_csv('dataset_project_1.csv', dtype=column_types)

# Write the modified DataFrame back to the CSV
df.to_csv('dataset_project_1.csv', index=False)
```

Considering the potential impact of geographical factors on funding outcomes, we will enhance the current dataset by integrating the *"offices.csv"* table with the merged dataset *"dataset_project_1.csv"*. This integration will incorporate geographical variables such as cities and states in the United States from "offices.csv" into the dataset called *"dataset_project_2.csv"*.

## 4.1 OLS Regression

In the initial project, the dependent variable Y is the amount of funds raised, focusing on how various factors influence startup fundraising success. Based on the preliminary data analysis, the relationship between Y and X (variables such as pre-money valuation) seems to be nonlinear. Economic theories such as the Law of Diminishing Returns suggest that after reaching a certain investment threshold, additional funding does not proportionally increase fundraising outcomes, supporting a nonlinear relationship. Additionally, the variability in how startups leverage different levels of investment points to complex dynamics that simple linear models may not capture.

For the initial four regression analyses, the independent variables chosen from the original dataset include 'pre_money_valuation', 'participants', 'funding_round_type', and 'fund_raised_amount'. These variables were selected to explore their direct impact on the fundraising outcomes based on their prominence in venture capital dynamics. The subsequent four regression analyses utilize variables such as 'team_size', 'no_founders', 'no_company_socials', and 'no_tags' sourced from the additional dataset. These variables are intended to examine the influence of operational and social engagement factors on fundraising effectiveness, providing a broader perspective on the determinants of startup success.

The preferred regression specification would be a model that includes interactions between key variables (such as pre-money valuation) and controls for potential confounders like market conditions and economic cycles. This model is preferred because it can reveal the combined effects of multiple factors on fundraising success, providing a deeper understanding of how these variables interplay in the venture capital context. 

```python
! pip install stargazer
from stargazer.stargazer import Stargazer
from IPython.core.display import HTML
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import statsmodels.api as sm
from statsmodels.iolib.summary2 import summary_col
from linearmodels.iv import IV2SLS
import warnings
warnings.filterwarnings("ignore")
```

For the initial regression analysis, we will focus solely on the categorical independent variable 'funding_round_type' and its associated dummy variables. This first regression serves as a foundation for the subsequent seven analyses. We will use the 'venture' category as our reference group because it contains the majority of observations. To simplify the analysis, we will combine less frequent categories such as 'private-equity', 'crowdfunding', and 'post-ipo' into the existing 'other' group. Consequently, we will break down the 'funding_round_type' into six distinct dummy variables to construct our initial regression model. This approach allows us to assess the impact of different funding types on the dependent variable while maintaining a clear and manageable model structure.

### Regression 1

```python
reg1 = sm.OLS(endog=df1['raised_amount_usd'], exog=df1[['const', 'angel', 'series-a', 'series-b', 'series-c+', 'other']], missing = 'drop').fit()
stargazer = Stargazer([reg1])
stargazer.custom_columns(["raised_amount_usd"],[1])
HTML(stargazer.render_html())
```

The first regression analysis is examining the impact of **funding round types** on the amount of capital raised, we observe that startups in angel funding rounds typically raise about \\$7.85 million less compared to those in the baseline venture rounds, which average a raise of approximately $81.60 million. Furthermore, startups in Series A rounds tend to raise \\$22.46 million less than venture rounds, while Series B rounds show an increase, with startups raising \\$31.85 million more. Notably, companies in Series C or higher rounds have a pronounced increase in raised funds, averaging \\$130.01 million more than the baseline venture rounds.

The R-squared value of 0.021 indicates that only 2.1% of the variability in the amount raised by startups is explained by the funding round types in this model. This suggests that while there are statistically significant differences between funding rounds, they do not account for most of the variation in the amount raised, pointing to other factors not included in this model that significantly influence fundraising outcomes. The F statistic is significant, suggesting that the model is statistically significant and that there is a relationship between the type of funding round and the amount raised. However, the low R-squared value indicates a need for caution in interpreting the results, as other unaccounted variables may also play a significant role in determining fundraising amounts. As the first model, it does not imply causation but rather an association between the type of funding round and fundraising outcomes.

The first regression explores the impact of different funding rounds on startup fundraising, providing insights into how maturity and perceived risk influence investment amounts. Regression one helps in understanding the variation in venture capital behavior and how different funding stages correlate with fundraising success.

The objective function of our regression tree is to ascertain the optimal number of terminal nodes that minimize the sum of squared differences between the actual fundraising amounts and the predicted values for each unique segment of startups. This is done by averaging the fundraising outcomes of startups within each segment, penalized by a factor of the tree's complexity parameter multiplied by the number of terminal nodes. This approach essentially aims to achieve the lowest mean squared error, balancing model accuracy with complexity.

In our context, the regularization parameter, typically denoted as Î±, modulates the balance between the tree's complexity and its fit to the training data. A higher value of Î± imposes a stiffer penalty for complexity, resulting in a more streamlined model with fewer branches, which mitigates the risk of overfitting and typically enhances the model's performance on unseen data. Conversely, a lower Î± value allows for a more intricate model that may fit the training data more closely but risks overfitting, which could degrade its performance on new data.
